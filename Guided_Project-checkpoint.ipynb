```python
# Create function to download url into a specified directory
import os
from urllib2 import urlopen, URLError, HTTPError

def downloadfile(url,download_dir):
    # Open the url
    try:
        f = urlopen(url)
        #print "Starting download of " + url

        # Open our local file for writing
        fname = os.path.join(download_dir,os.path.basename(url))
        with open(fname, "wb") as local_file:
            local_file.write(f.read())
            
        #print "Finished downloading " + url

    #handle errors
    except HTTPError, e:
        print "HTTP Error:", e.code, url
    except URLError, e:
        print "URL Error:", e.reason, url
```

```python
# Download the csv files containing the list of players and list of teams
# These would be used in next step to download the player and team information in the next step
# These url are used by the search function on the website to lookup players and teams
# These discovered by examining the website in developer tools
player_list_url = "https://d2cwpp38twqe55.cloudfront.net/short/inc/players_search_list.csv"
team_list_url = "https://d2cwpp38twqe55.cloudfront.net/short/inc/teams_search_list.csv"
coaches_list_url = "https://d2cwpp38twqe55.cloudfront.net/short/inc/coaches_search_list.csv"
executives_list_url = "https://d2cwpp38twqe55.cloudfront.net/short/inc/executives_search_list.csv"
downloadfile(player_list_url,os.getcwd())
downloadfile(team_list_url,os.getcwd())
downloadfile(coaches_list_url,os.getcwd())
downloadfile(executives_list_url,os.getcwd())
```

```python
# Load the player list
import pandas as pandas
import numpy as np
player_list = pandas.read_csv(os.path.basename(player_list_url),header=None,names=("Player_ID","Player_Name","Seasons_Active","Active"),
                              dtype = {"Player_ID":np.dtype('U'),"Player_Name":np.dtype('U'),"Seasons_Active":np.dtype('U'),"Active":bool})
player_list_active_seasons = player_list.Seasons_Active.str.split("-",expand=True)
player_list_active_seasons.columns = ["From_Season","To_Season"]
player_list_active_seasons = player_list.Seasons_Active.str.split("-",expand=True)
player_list_active_seasons.columns = ["From_Season","To_Season"]
player_list_active_seasons.From_Season=pandas.to_numeric(player_list_active_seasons.From_Season,errors='coerce').fillna(0).astype(np.int64)
player_list_active_seasons.To_Season=pandas.to_numeric(player_list_active_seasons.To_Season,errors='coerce').fillna(0).astype(np.int64)
player_list = player_list.join(player_list_active_seasons)

# Filter the player list to only the players that have played in any game since 2003 or those currently active
filt = (player_list.Active) | (player_list.From_Season >= 2003) | (player_list.To_Season >= 2003)
player_list_filtered = player_list[filt]
```

```python
import time
# Now that we have filtered the players we want 
# We can proceed with downloading and saving the html files on all the players
# We have discovered that the URL specification for each player is given by
# url = https://www.basketball-reference.com/players/{first_character_of_player_id}/{player_id}.html
base_url = "https://www.basketball-reference.com/players/" # Base url for all the players
player_folder = "players"
for i in player_list_filtered.Player_ID:
    first_char = i[0]
    html = i + ".html"
    url = os.path.join(base_url,first_char,html) # full url for the html file containing the player info
    downloadfile(url,player_folder)
    time.sleep(1) # Sleep for 1 second before download the next file. To be nice to the website
```
